#### Exercise 1

Write a factorial function

```python
def factorial(n):
    i = 1
    for x in range(n):
        i = (x+1)*i
    return i
```

```python
factorial(6)
```

#### Exercise 2
write a binomial variable function

```python
from numpy.random import uniform as uni
def binomial_rv(n,p):
    count = 0
    for i in range(n):
        if uni()<p:
            count+=1
        else:
            pass
    return count
```

```python
binomial_rv(10,.9)
```

#### Exercise 3

Estimate $\pi$ with a monte carlo simulation (Needed help on this one, though I understood the basics. So I redid it but with volume instead of area, to at least show that now I know what was going on)


```python
def pi_mc(n):
    count = 0
    for i in range(n):
        u,v = uni(), uni()
        d = np.sqrt((u-.5)**2 + (v-.5)**2)
        if d < .5:
            count +=1
    area_est = count/n
    print(area_est*4)
    
```

```python
#pi_mc(1000000000)
```

```python
def pi_mc2(n):
    count = 0
    for i in range(n):
        u,v,w = uni(), uni(), uni()
        d = np.sqrt((u-.5)**2+(v-.5)**2+(w-.5)**2)
        if d< .5:
            count +=1
    vol_est = count/n
    print(vol_est*6)
```

```python
n = 100000
print(pi_mc(n))
print(pi_mc2(n))
```

#### Exercise 4

The coin flipping machine (did it in different way, but I think it still ends up doing the same thing)

```python
def gambler(n):
    count = 0
    for x in range(n):
        if count != 3:
            i = uni()
            if i > .5:
                count +=1
            else:
                count = 0
        else:
            break
    if count == 3:
        print('you won one dollar')
        print(x)
    else:
        print('you lose')
        print(x)
```

```python
gambler(10)
```

#### Exercise 5

simulate and plot a time series (did this more or less without help, just looking once to see how they started (and learned it was a list))

```python
def ts_plot(t,α):
    x = []
    t0 = 0
    for i in range(t):
        t0 = α*t0 + np.random.randn()
        x.append(t0)
    plt.plot(x, label="α = " + str(α))
    plt.legend()
```

```python
ts_plot(200,.9)
```
